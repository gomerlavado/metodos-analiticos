# 1.2 Potencia compleja

```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega = 2 * math.pi * 60
time = np.linspace(0, 0.1, 1000)
```

```python
irms = 1
im = math.sqrt(2) * irms
thetai = math.radians(30)
```

```python
l = 0.5
z = complex(0, omega*l)
```

```python
it = im * np.cos(omega * time + thetai)
vt = - math.sqrt(2) * omega * l * irms * np.sin(omega * time + thetai)
```

```python
s = z * irms**2
```

```python
# Potencia reactiva
q = s.imag
print(q)
```

```python
plt.plot(time, vt*it, "tab:blue", time, it, "tab:orange", time, vt, "tab:purple")
plt.hlines(q, 0, 0.1, colors="tab:blue", linestyles="dashed")
plt.xlabel("Time [s]")
#plt.savefig('reactive-power-l.png', dpi=1000)
plt.show()
```

```python
vrms = 220
vm = math.sqrt(2) * vrms
thetav = math.radians(60)
```

```python
c = 0.0005
z = complex(0, -1/(omega*c))
```

```python
vt = vm * np.cos(omega * time + thetav)
it = - math.sqrt(2) * omega * c * vrms * np.sin(omega * time + thetav)
```

```python
s = vrms ** 2 / z.conjugate()
q = s.imag
print(q)
```

```python
plt.plot(time, vt*it, "tab:blue", time, it, "tab:orange", time, vt, "tab:purple")
plt.hlines(q, 0, 0.1, colors="tab:blue", linestyles="dashed")
plt.xlabel("Time [s]")
#plt.savefig('reactive-power-c.png', dpi=1000)
plt.show()
```
