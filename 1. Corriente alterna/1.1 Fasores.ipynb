# 1.1 Fasores

```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Single phase system
vrms = 220
irms = 1
```

```python
# max voltage and current
vm = math.sqrt(2) * vrms
im = math.sqrt(2) * irms
thetav = math.radians(60)
thetai = math.radians(30)
phi = thetav - thetai
omega = 2 * math.pi * 60
time = np.linspace(0, 0.1, 1000)
```

```python
vt = vm * np.cos(omega * time + thetav)
it = im * np.cos(omega * time + thetai)
```

```python
plt.plot(time, vt, "tab:purple", time, it, "tab:orange", time, vt*it, "tab:blue")
plt.hlines(220, 0, 0.1, colors="tab:purple", linestyles="dashed")
plt.xlabel("Time [s]")
# plt.savefig('sinusoids.png', dpi=1000)
plt.show()
```

```python
pf = math.cos(phi)
pavg = vrms * irms * pf
print(pavg)
```

```python
vz = complex(vrms * math.cos(thetav), vrms * math.sin(thetav))
iz = complex(irms * math.cos(thetai), irms * math.sin(thetai))
```

```python
s = vz * iz.conjugate()
print(s)
print(s.real)
print(cmath.polar(s))
```
